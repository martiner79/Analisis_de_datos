```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tabla = pd.read_csv("finance_data.csv")
tabla
```

```python
tabla_ch = tabla[tabla.Artículo == 'Chocolate']
tabla_ch.head()
```

```python
info = tabla[tabla.Categoría == 'Hogar']
info.Artículo.value_counts()
```

```python
info.Artículo.value_counts().plot(x='Artículo', y='Cantidad',figsize=(10,5), color='red');
plt.plot(3,24,"o");
```

```python
#Tipos de gr+afocps

#Gráfico de Barras verticales.
info.Artículo.value_counts().plot(kind='bar');
```

```python
#Barras con colores
productos = pd.read_csv("productos.txt")
sns.barplot(x='Productos', y='Ganancias', data=productos);
```

```python
#1- Gráfico de Barras horizontales.
info.Artículo.value_counts().plot(kind='barh');
```

```python
#1- Gráfico de Pastel
info.Artículo.value_counts().plot(kind='pie');
```

```python
#1- Gráficos de Coordenadas.
sns.scatterplot(x='Unidades', y='Ganancia', data=tabla);
```
